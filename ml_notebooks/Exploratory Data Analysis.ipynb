```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bestsellers_data = pd.read_csv("C:\\Users\\mzalendo_ke\\Documents\Documents\\Machine Learning\\bestsellers with categories.csv")
```

```python
# Display basic info and summary statistics
print("Dataset Info:")
print(bestsellers_data.info())
print("\nDataset Summary:")
print(bestsellers_data.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(bestsellers_data.isnull().sum())
```

```python
# Create subplots for distributions of User Rating, Reviews, and Price
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
```

```python
sns.histplot(bestsellers_data['User Rating'], bins=10, kde=True, ax=axes[0])
axes[0].set_title('User Rating Distribution')
```

```python
sns.histplot(bestsellers_data['Reviews'], bins=20, kde=True, ax=axes[1])
axes[1].set_title('Reviews Distribution')
```

```python
sns.histplot(bestsellers_data['Price'], bins=20, kde=True, ax=axes[2])
axes[2].set_title('Price Distribution')

plt.tight_layout()
plt.show()
```

```python
corr_matrix = bestsellers_data[['User Rating', 'Reviews', 'Price']].corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Genre', data=bestsellers_data)
plt.title('Distribution of Fiction vs Non-Fiction')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Genre', y='User Rating', data=bestsellers_data)
plt.title('User Rating by Genre')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Genre', y='Reviews', data=bestsellers_data)
plt.title('Reviews by Genre')
plt.show()
```
