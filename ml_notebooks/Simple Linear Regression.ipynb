```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pnt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df_sal = pd.read_csv("C:\\Users\\Student\\Downloads\\Salary_Data.csv")
```

```python
df_sal.head(10)
```

```python
df_sal.describe()
```

```python
pnt.scatter(df_sal["YearsExperience"],df_sal["Salary"], color='lightcoral')
pnt.title("Salary vs YearsExperience")
pnt.xlabel('YearsExperience')
pnt.ylabel('Salary')
pnt.show()
```

```python
x=df_sal.iloc[:, :1]
y=df_sal.iloc[:, 1:]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
regresser=LinearRegression()
regresser.fit(X_train, Y_train)
```

```python
y_pread_test=regresser.predict(X_test)
y_pread_train=regresser.predict(X_train)
```

```python
pnt.scatter(X_train, Y_train, color='lightcoral')
pnt.plot(X_train, y_pread_train)
pnt.title('Salary vs Experience of training set' )
pnt.legend(['X_train/pread(Y_test)','X_train/Y_train'], title='sal/Exp', loc='best', facecolor='white')
pnt.xlabel("YearsExperience")
pnt.ylabel("Salary")
```

```python
pnt.scatter(X_test, Y_test, color='lightcoral')
pnt.plot(X_train, y_pread_train)
pnt.title('Salary vs Experience of test set' )
pnt.legend(['X_train/pread(Y_test)','X_train/Y_train'], title='sal/Exp', loc='best', facecolor='white')
pnt.xlabel("YearsExperience")
pnt.ylabel("Salary")
```

```python
print(f'Coefficients: {regresser.coef_}')
```
