```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
dataset=pd.read_csv("C:\\Users\\Student\\Downloads\\advertising.csv")
```

```python
dataset.head()
```

```python
x=dataset[['TV', 'Radio', 'Newspaper']]
y=dataset['Sales']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

```python
mlr=LinearRegression()
mlr.fit(X_train, Y_train)
```

```python
print(f'intercept: {mlr.intercept_}')
```

```python
print(f'coefficient: {mlr.coef_}')
```

```python
print('coefficient')
list(zip(x, mlr.coef_))
```

```python
y_pread_mlr=mlr.predict(X_test)
```

```python
Y_pred_test = mlr.predict(X_test)
Y_pred_train = mlr.predict(X_train)
print('Prediction for test set: {}'.format(Y_pred_test))
```
